```python
# sequential model banaya hai

model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  # input mai 10 features hai and output hai 64 neurons
    layers.Dense(32, activation='relu'),                                   # input mai 64 neurons hai and than output mai 32 neurons hai
    layers.Dense(y_train.shape[1], activation='sigmoid')                   # input mai 32 neurons hai and than output mai number of unique dishes hai
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# 'binary_crossentropy' is used for binary classification problems, where the target variable has two classes (e.g., 0 or 1, True or False).
#
```

```python
# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```
